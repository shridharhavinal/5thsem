```python
from pandas import DataFrame
df1 = DataFrame(
              [
                  ['Shridhar','Havinal','Full-time','administration',2,20000],
                  ['Omkar','Sharanarthi','intern','technical',1,5000],
                  ['Rahul','Badakuri','Full-time','administration',5,10000],
                  ['Shivanand','Pujari','part-time','technical',3,10000],
                  ['Santosh','Hanjagi','Full-time','Management',6,20000]
              ]
    ,columns=['First Name','Last Name','Type','Department','YoE','salary']
              )
df1
```

```python
import pandas as pd
table=pd.pivot_table(df1,index='Type')
table
```

```python
table=pd.pivot_table(df1,index='Type',values='salary')
table
```

# Melt Function

```python
import pandas as pd
df = pd.DataFrame(
              [
                  ['Rahul',19,14,18,19],
                  ['Santosh',18,10,17,16],
                  ['Omkar',17,16,18,15]
              ]
    ,columns=['Name','OSA','JAVA','SE','DS']
             )
print(df)
df.melt(id_vars=['Name'])
```
