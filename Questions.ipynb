```python
from pandas import DataFrame
df1 = DataFrame(
              [
                  ['Shridhar','Havinal','Full-time','administration',2,20000],
                  ['Omkar','Sharanarthi','intern','technical',1,5000],
                  ['Rahul','Badakuri','Full-time','administration',5,10000],
                  ['Shivanand','Pujari','part-time','technical',3,10000],
                  ['Santosh','Hanjagi','Full-time','Management',6,20000]
              ]
    ,columns=['First Name','Last Name','Type','Department','YoE','salary']
              )
df1
```

Make a pivot table which shows average salary of each type of employee for each 
department

```python
import pandas as pd
table=pd.pivot_table(df1,index='Type',columns='Department',values='salary')
table
```

Make a pivot table which shows the sum and mean of the salaries of each type of 
employee and the number of employees of each type

```python
table=pd.pivot_table(df1,index='Type',values='salary',aggfunc=['mean','sum','count'])
table
```
